# **Feature Selection**

```python
## To load csv fiile from local device to google colab
from google.colab import files
uploaded = files.upload()
```

```python
#Importing all the library necessary for analysis and viewing dataset
import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv(io.BytesIO(uploaded['X_train.csv']))
dataset
```

```python
#for feature selection
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

#drop dependent feature from dataset
X_train = dataset.drop(['Id','SalePrice'],axis=1)
X_train
```

```python
#Capture the output/dependent feature
y_train = dataset[['SalePrice']]
y_train
```

```python
#First specify the lasso regression model 
#select a suitable alpha value
#the bigger the alpha less feature will be selected

feature_sel_model = SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(X_train,y_train)
```

```python
#this will tell you that which feature is important for model prediction 
feature_sel_model.get_support()
```

```python
#this is how we make the list of selected feature
selected_feature = X_train.columns[feature_sel_model.get_support()]

#print the total number of selected and total number of features
print('total number of features are : {}'.format((X_train.shape[1])))
print('total number of selected features are : {}'.format((len(selected_feature))))
```

```python
X_train = dataset[selected_feature]
```
