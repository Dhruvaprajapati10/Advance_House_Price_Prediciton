```python
## To load csv fiile from local device to google colab
from google.colab import files
uploaded = files.upload()
```

```python
## Importing all the library necessary for analysis and viewing dataset
import io
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv(io.BytesIO(uploaded['train.csv']))
dataset
```

```python
dataset.shape
```

```python
#Handling the missing value of categorical feature
feature_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype == 'O']
for feature in feature_nan:
  print('{} --> {} Missing Values '.format(feature,np.sum(dataset[feature].isnull())))
```

```python
dataset[feature_nan].head()
```

```python
for feature in feature_nan:
  data = dataset.copy()
  data[feature] = data[feature].fillna('Missing')
  dataset[feature] = data[feature]
  
```

```python
dataset[feature_nan].isnull().sum()
```

```python
dataset.head()
```

```python
#Handling the missing value of numerical feature
numeric_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype != 'O']
for feature in numeric_with_nan:
  print('{} --> {} Missing Values '.format(feature,np.sum(dataset[feature].isnull())))
```

```python
dataset[numeric_with_nan].head()
```

```python
for feature in numeric_with_nan:
  median_value = dataset[feature].median()

  dataset[feature+'nan'] = np.where(dataset[feature].isnull(),1,0)
  dataset[feature].fillna(median_value,inplace=True)

dataset[numeric_with_nan].isnull().sum()
```

```python
dataset.head()
```

```python
# Temporal Variable

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
  dataset[feature] = dataset['YrSold'] - dataset[feature]
```

```python
dataset.head()
```

```python
num_feature = ['LotFrontage','LotArea','GrLivArea','SalePrice','1stFlrSF']

for feature in num_feature:
  dataset[feature] = np.log(dataset[feature])
  
```

```python
dataset.head()
```

```python
#Handling Rare categorical feature
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
categorical_feature
```

```python
for feature in categorical_feature:
  temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
  temp_df = temp[temp>0.01].index
  dataset[feature] = np.where(dataset[feature].isin(temp_df),data[feature],'Rare_var')
```

```python
dataset.head(100)
```

```python
mszoning = [feature for feature in np.unique(dataset.MSZoning) ]
mszoning
```

```python
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O' ]
from sklearn.preprocessing import LabelEncoder 
for feature in categorical_feature:
  label = LabelEncoder()
  dataset[feature] = label.fit_transform(dataset[feature])
```

```python
dataset.head()
```

# **Feature** **Scaling**

```python
feature_scale = [feature for feature in dataset.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scalar.fit(dataset[feature_scale])
```

```python
dataset.head()
```

```python
data = pd.concat([dataset[['Id','SalePrice']],
                 pd.DataFrame(scalar.transform(dataset[feature_scale]),columns=feature_scale)],axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv',index=False)
```
